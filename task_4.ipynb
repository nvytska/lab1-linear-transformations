```python
from task_3 import read_off, plot_diff3d
import numpy as np
```

```python
def run_task4(path_to_off, theta_xy, theta_yz, theta_xz):
    print("Task 4")
    verts, faces = read_off(path_to_off)
    Rxy = np.array([[np.cos(theta_xy), -np.sin(theta_xy), 0],[np.sin(theta_xy),  np.cos(theta_xy), 0],[0, 0, 1]])

    Ryz = np.array([[1, 0, 0],[0, np.cos(theta_yz), -np.sin(theta_yz)],[0, np.sin(theta_yz),  np.cos(theta_yz)]])

    Rxz = np.array([[ np.cos(theta_xz), 0, np.sin(theta_xz)],[0, 1, 0],[-np.sin(theta_xz), 0, np.cos(theta_xz)]])

    R_total = Rxz @ Ryz @ Rxy
    print("Combined rotation matrix (R_total = R_xz 路 R_yz 路 R_xy):\n", R_total)
    rotated = verts @ R_total.T
    plot_diff3d(verts, rotated, faces,"Task 4: Combined rotation ")
```
