```python
def run_task2(X, a, b, theta):
    print("Task 2")
    A_stretch = np.array([[a,0], [0, b]])
    A_shear  = np.array([[1, a], [b, 1]])
    A_rot  = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

    comminations = {
        "Stretch - shear - rotate" : A_stretch @ A_shear @ A_rot,
        "Shear - stretch - rotate" : A_shear @ A_stretch @ A_rot,
        "Rotate - shear - stretch":  A_rot @ A_shear @ A_stretch
    }

    for c, A in comminations.items():
        X_new = X @ A.T
        plot_diff(X, X_new, f"{c}")
```
